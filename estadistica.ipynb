```python
import pandas as pd
import numpy as np
import plotly.express as px
from scipy.stats import f_oneway
```

```python
data_frame = pd.read_csv('leyes.csv', sep=';')
data_frame
```

## Visualización de datos de Fierro

```python
continuous = 'Fe (%)'
figure = px.scatter_3d(data_frame, x='Este (m)', y='Norte (m)', z='Cota (m)', color=continuous, color_continuous_scale=px.colors.sequential.Jet, range_color=[0.0, data_frame[continuous].quantile(0.95)])
figure.update_traces(marker=dict(size=5.0))
figure.show()
```


## Visualización de datos de Molibdeno

```python
continuous = 'CuS (%)'
figure = px.scatter_3d(data_frame, x='Este (m)', y='Norte (m)', z='Cota (m)', color=continuous, color_continuous_scale=px.colors.sequential.Jet, range_color=[0.0, data_frame[continuous].quantile(0.95)])
figure.update_traces(marker=dict(size=5.0))
figure.show()
```

## Eliminación de datos outliers

```python
data_frame['Fe (%)'][data_frame['Fe (%)'] <= 0] = np.nan
data_frame['Mo (%)'][data_frame['Mo (%)'] <= 0] = np.nan
data_frame['As (%)'][data_frame['As (%)'] <= 0] = np.nan
data_frame['Zn (%)'][data_frame['Zn (%)'] <= 0] = np.nan
data_frame['CuS (%)'][data_frame['CuS (%)'] <= 0] = np.nan
```

```python
data_frame.describe()
```

```python
data_frame.describe()[['Fe (%)', 'Mo (%)', 'As (%)', 'Zn (%)', 'CuS (%)']].T
```

```python
stats = data_frame.describe()[['Fe (%)', 'Mo (%)', 'As (%)', 'Zn (%)', 'CuS (%)']].T
stats['IQR'] = stats['75%'] - stats['25%']
stats['95%'] = data_frame.select_dtypes(include='number').quantile(0.95)
stats
```

## Covarianza

```python
covariance = data_frame[['Fe (%)', 'Mo (%)', 'As (%)', 'Zn (%)', 'CuS (%)']].cov()
covariance
```

## ANOVA

```python
data_frame['Fe (%)'] = pd.to_numeric(data_frame['Fe (%)'], errors='coerce')
data_frame['Mo (%)'] = pd.to_numeric(data_frame['Mo (%)'], errors='coerce')
data_frame['As (%)'] = pd.to_numeric(data_frame['As (%)'], errors='coerce')
data_frame['Zn (%)'] = pd.to_numeric(data_frame['Zn (%)'], errors='coerce')
data_frame['CuS (%)'] = pd.to_numeric(data_frame['CuS (%)'], errors='coerce')

data_frame.dropna(subset=['Fe (%)', 'Mo (%)', 'As (%)', 'Zn (%)', 'CuS (%)'], inplace=True)

r_anova = f_oneway(data_frame['Fe (%)'], data_frame['Mo (%)'], data_frame['As (%)'], data_frame['Zn (%)'], data_frame['CuS (%)'])
r_anova
```
